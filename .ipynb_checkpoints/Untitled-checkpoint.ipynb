```python
import random

A=[random.randint(-5,5) for i in range(10)]
```

```python
while True:
    last_mas=A.copy()
    for i in range(0,len(A)):
        if A[i]!=abs(A[i]) and A[i+1]==min(A):
            A.remove(A[i])
            continue

    if last_mas!=A:
        last_mas=A.copy()
    else:
        break
    
print(A)
```
